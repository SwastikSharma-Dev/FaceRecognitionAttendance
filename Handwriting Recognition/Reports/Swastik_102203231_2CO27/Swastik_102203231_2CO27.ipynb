```python
# SWASTIK SHARMA 102203231
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

train_df = pd.read_csv("data.csv")
test_df = pd.read_csv("test.csv")

y_train = train_df['label']
x_train = train_df.drop('label', axis=1)
x_test = test_df

test_sizes = [0.4, 0.3, 0.25, 0.2, 0.1, 0.05]

k_values = [2, 4, 5, 6, 7, 10]

results = {}

for test_size in test_sizes: 
    x_train_split, x_val, y_train_split, y_val = train_test_split(x_train, y_train, test_size=test_size, random_state=40)
    
    scaler = StandardScaler()
    scaler.fit(x_train_split)
    x_train_split = scaler.transform(x_train_split)
    x_val = scaler.transform(x_val)
    x_test_scaled = scaler.transform(x_test)
    
    for k in k_values:
        classifier = KNeighborsClassifier(n_neighbors=k)
        classifier.fit(x_train_split, y_train_split)
        
        y_pred_val = classifier.predict(x_val)
        
        accuracy_val = accuracy_score(y_val, y_pred_val)
        cm = confusion_matrix(y_val, y_pred_val)
        
        results[(test_size, k)] = {'Validation Accuracy': accuracy_val, 'Confusion Matrix': cm}

    for params, metrics in results.items():
        print("Test Size: {}\t K: {}\n".format(params[0], params[1]))
        print("Validation Accuracy: {}\n".format(metrics['Validation Accuracy']))
        print("Confusion Matrix:\n{}\n".format(metrics['Confusion Matrix']))
        print("-----------------------------------\n")
```
